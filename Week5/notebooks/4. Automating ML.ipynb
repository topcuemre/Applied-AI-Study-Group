## LazyPredict

LazyPredict is a very simple library helping testing many classical machine learning classification or regression methods to be tested at once. It is especially helpful to get an insight on the overall performance of the dataset.

**Please note:** You may be required to install the library in a separate virtual environment to resolve dependency issues.

```python
# !pip install lazypredict
# # or
# !pip install git+https://github.com/shankarpandala/lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()
x, y = data.data, data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, shuffle=True, random_state=43)

classifier = LazyClassifier()
models, predictions = classifier.fit(x_train, x_test, y_train, y_test)

models
```
